```python
#plotting the graph b/w Category and Sales
plt.figure(figsize=(8,8))
plt.bar('Category', 'Sales', data=df)
plt.title('Category Vs Sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

From the above graph the following inferences can be made - 
1) Most goods sold are from the Technology Category

2) The least sold goods belong to the Furniture Cateogry.

```python
#plotting the graph b/w Sub-Category and Sales
plt.figure(figsize=(20,10))
plt.bar('Sub-Category', 'Sales', data=df)
plt.title('Sub-Category Vs Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.show()
```


If we look at the sales of sub-categories, we infer that the least sold quantities are Labels, Furnishings, Art, Paper,
Envelopes and Fasteners. The most sold quantities are Binders, Supplies, Machines and Copiers. The remaining have a moderate
sales.

```python
df['State'].value_counts() #checking the count of each State in the dataset
```

```python
#countplot for the different states in the dataset
plt.figure(figsize=(20,20))
sns.countplot(x=df['State'])
plt.xticks(rotation=90)
plt.title("State")
plt.show()
```

As you can see, the most number of sales have been recorded in California, followed by New York and Texas. States such as Nebraska, Iowa, New Mexico, Nevada, Montana, New Hampshire, Wyoming, etc have very low sales. The retail at these states must be significantly improved for better profits.

```python
#pie chart depicting the profit made through shipments
Shipping=df.groupby(['Ship Mode'])[['Quantity','Profit']].mean()
Shipping.plot.pie(subplots=True, figsize=(15,15), autopct='%1.1f%%', labels=Shipping.index)
```

From the Ship Mode - Quantity pie chart we can infer that most goods are shipped through Standard Class, followed by Second Class. 
From the Ship Mode - Profit pie chart it is evident that even though the quantity shipped thorugh the First Class are comparitively low it is the most profitable.

```python
#paiplot between the correlated quantities
figsize=(15,10)
sns.pairplot(df, hue='Region')
plt.show()
```

```python

sns.kdeplot(df['Profit'], label='PROFIT', color='red', fill=True)
sns.kdeplot(df['Sales'], label='SALES', color='green', fill=True)
plt.xlim(-400,1000)
plt.title('Sales vs Profits')
plt.legend()
plt.show()
```

```python
#heatmap of the correlated quantities
sns.heatmap(df.corr(),linewidth=0.5, annot=True, cmap='coolwarm')
plt.show()
```

From the above Heatmap we can infer the following :

1) Sales and Profit are positively Correlated (Sales increases, profit increases)

2) Quantity and Sales are positvely Correlated (Quantity increases, Sales increases)

3) Discount and Profit are negatiely Correlated (More Discount, less Profit)



```python
#bar graph depicting profit made in specific regions
plt.figure(figsize=(6,8))
sns.barplot(data=df, x='Region', y='Profit', hue='Segment', palette='hsv')
plt.show()
```

From the above bar graph we see that - 
1) The West region has the highest number of Consumer and Corporate Segments.

2) Central region has the least number of Consumers.


3) The East region has the highest Home Office Segments.

```python
#lineplot of Profit vs Quantity
plt.figure(figsize=(10,4))
sns.lineplot('Quantity', 'Profit', data=df, color='y', label='Profit vs Quantity')
plt.legend()
```

This is the lineplot for Profit vs Discount. 
Clearly as more quantities are sold, profit increases.

```python
#lineplot of Profit vs Discount
plt.figure(figsize=(10,4))
sns.lineplot('Discount', 'Profit', data=df, color='b', label='Profit vs Discount')
plt.legend()
```

It is evident that giving discounts leads to loss, but without doing so the Superstore might lose their customers.
